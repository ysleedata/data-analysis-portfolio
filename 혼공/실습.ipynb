```python
n = int(input("입력: "))
result = []
for i in range(1, n + 1):
    if n % i == 0:
        result.append(i)
print("약수:",result)
```

```python
ans = 9

while True:
    num = int(input("정수를 입력하세요(1~10): "))
    if ans == num:
        print("정답입니다!")
        break
    else:
        print("틀렸습니다. 다시 입력하세요.")
```
