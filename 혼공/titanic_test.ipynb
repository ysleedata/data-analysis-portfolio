```python
import pandas as pd

url = "https://gist.githubusercontent.com/jwalsh/ce1dc0436aba5b7a5c9666f47fa5a380/raw/5ce3854392b43ff97907112d344fc008229b0445/titanic.csv"

df = pd.read_csv(url)
# df = pd.read_csv("data/train.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending = False)
# isnull은 비어있는 칸을 True라고 셈.
# sum은 컬럼별로 True를 다 더함.
# sort_values(ascending = False)는 구한 개수를 큰 숫자부터 작은 숫자 순서로 정렬
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
# 최빈값이 여러 개일 수 있어서 판다스에서는 리스트로 나옴. 그 중 첫번째 값을 꺼내기 위해선 [0]이 필요.
```
