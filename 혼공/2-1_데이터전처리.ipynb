**난이도 하 (기초 전처리 — 분석 가능 형태 만들기)**

```python
import pandas as pd

data = {
    "user_id": [101, 102, 103, 104, 105],
    "age": [23, None, 31, None, 28],
    "gender": ["F", "M", None, "F", "M"],
    "signup_date": ["2024-01-03", "2024-01-05", "2024-01-07", "2024-01-10", "2024-01-12"]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
df2 = df.dropna(subset=['age'])
df2
```

```python
df2['signup_date'] = pd.to_datetime(df['signup_date'])
df2.info()
```

```python
df2['age_group'] = (df2['age']//10*10)
df2['age_group'] = df2['age_group'].astype(int).astype(str) + "대"
age_analysis = df2.groupby('age_group')['user_id'].nunique()

age_analysis
```

**난이도 중 (정제 + 파생변수 생성)**

```python
data = {
    "order_id": [1, 2, 3, 4, 5, 6],
    "user_id": [101, 101, 102, 103, 103, 103],
    "price": [12000, 15000, None, 8000, 9000, None],
    "category": ["food", "food", "beauty", "food", None, "beauty"]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['price'] = df['price'].fillna(df['price'].mean())

df
```

```python
df['category'] = df['category'].fillna('unknown')

df
```

```python
user_analysis = df.groupby('user_id')['price'].agg(['sum','mean'])
user_analysis.columns = ['총 구매금액','평균 구매금액']

user_analysis
```

**난이도 상 (실무형 전처리 — 분석용 테이블 완성)**

```python
data = {
    "user_id": [1, 2, 3, 4, 5],
    "join_date": ["2023-12-01", "2024-01-15", "2024-01-20", "2024-02-05", "2024-02-20"],
    "last_login": ["2024-02-28", None, "2024-02-25", "2024-02-27", None],
    "total_payment": [120000, 0, 45000, 89000, 0]
}

df = pd.DataFrame(data)
df
```

```python
df['join_date']=pd.to_datetime(df['join_date'])
df.info()
```

```python
# 날짜 데이터를 파이썬이 이해할 수 있도록 바꿔주기
df['last_login'] = pd.to_datetime(df['last_login'])

# 활동 여부 만들기
df.loc[df['last_login'].isnull(),'활동여부']='Inactive'
df.loc[df['last_login'].notnull(),'활동여부']='Activate'
df
```

```python
# 결제여부 만들기
df.loc[df['total_payment'] > 0,'결제여부']='Yes'
df.loc[df['total_payment']== 0,'결제여부']='No'
df
```

```python
# 최근 접속 일수 (로그인 안 한 사람은 0으로 표시)
today = pd.to_datetime('2026-01-21')

df['최근 접속 일수'] = (today - df['last_login']).dt.days
df['최근 접속 일수'] = df['최근 접속 일수'].fillna(0)
df
```
