### 문제 9. 불리언 필터로 `자치구명 == '강남구'` AND `건물면적(㎡) >= 85`인 행만 선택하고 `['자치구명','법정동명','건물면적(㎡)','물건금액(만원)']`만 보세요.

```python
mask = (df['자치구명'] == '강남구') & (df['건물면적(㎡)'] >= 85)
seoul2024_gangnam_85 = df.loc[mask, ['자치구명','법정동명','건물면적(㎡)','물건금액(만원)']]
```
