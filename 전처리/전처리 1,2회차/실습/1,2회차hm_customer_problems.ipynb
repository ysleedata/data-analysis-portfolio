```python
conditions = (df['Active'] == 1) & (df['fashion_news_frequency'] == 'Regularly')

df.loc[conditions, ['customer_id','fashion_news_frequency','age']]
```

```python
# culb_member_status가 Active 이거나 FN이 0이고 age가 35세 이상인 행

conditions = (df['club_member_status'] == 'Active') | (df['FN'] == 0) & (df['age'] >= 35)

result = df.loc[conditions, ['customer_id','fashion_news_frequency','age']]
result.head()
type(conditions)
```
