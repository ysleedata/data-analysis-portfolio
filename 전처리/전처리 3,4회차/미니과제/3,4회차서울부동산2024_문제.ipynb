### 문제 1 — 조건 필터링(AND)

```python
# TODO: 여기에서 코드를 작성하세요.
df_gangnam_of = (df['자치구명'] == "강남구") & (df['건물용도'] == "오피스텔")

df.loc[df_gangnam_of]
```

### 문제 2 — 조건 필터링(OR)

```python
# TODO: 여기에서 코드를 작성하세요.
condition = (df['층'] >= 10) | (df['물건금액(만원)'] >= 100000)
df.loc[condition]
```

### 문제 10 — 결측치 채우기(수치형)

### 문제 10-2 — 결측치 채우기(수치형)

```python
# TODO: 여기에서 코드를 작성하세요.
df_new = df.copy()

df_new['건축년도'] = df_new['건축년도'].fillna(
    df_new.groupby('자치구명')['건축년도'].median()
)
```
