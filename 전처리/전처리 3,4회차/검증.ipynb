```python
import pandas as pd
import numpy as np
dup_df = pd.DataFrame({
    "PassengerId": [1, 1, 2, 2, 2, 3, 3],
    "Name":        ["Allen, Mr.", "Allen, Mr.", "Baker, Mr.", "Baker, Mr.", "Baker, Mr.", "Clark, Mrs.", "Clark, Mrs."],
    "Age":         [22, 22, np.nan, 35, 35, 29, np.nan],
    "Fare":        [7.25, 7.25, 8.05, np.nan, 8.05, 71.28, 71.28],
    "Embarked":    ["S", "S", "S", "S", "S", "C", "C"],
})
dup_df
```

```python
temp = dup_df.copy()
```

```python
temp.isna().sum(axis=1)
```

```python
temp['na_count'] = temp.isna().sum(axis=1)
temp
```

```python
best_rows = (
    temp.sort_values(["PassengerId", "na_count"])   # 결측 적은 행이 위로
       .drop_duplicates(subset=["PassengerId"], keep="first")  # PassengerId별 1개만 남김
       .drop(columns=["na_count"])
)

best_rows
```
